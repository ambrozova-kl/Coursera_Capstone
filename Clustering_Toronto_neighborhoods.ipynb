At first some libraries and packages must be imported.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

Let´s get the text of the website.

```python
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

The table is of class "wikitable sortable", so let´s extract that and read it to pandas dataframe.

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
df = pd.read_html(str(My_table))[0]
```

Checking how many neighbourhoods are not assigned.

```python
df[df.Neighbourhood == 'Not assigned']
```

Remove the rows with a borough not assigned.

```python
Assig_bor_only = df[df.Borough != 'Not assigned']
Assig_bor_only
```

How many neighbourhood names are still not assigned?

```python
Assig_bor_only[Assig_bor_only.Neighbourhood == 'Not assigned']
```

Only one neigbourhoods name needs to be replaced.

```python
Assig_neg_only1 = df[df.Neighbourhood != 'Not assigned']
Assig_neg_only2 = Assig_neg_only1.append({'Postcode' : 'M7A' , 'Borough' : 'Queen\'s Park', 'Neighbourhood' : 'Queen\'s Park'} , ignore_index=True)
Assig_neg_only2
```

Now just merge rows with the same postcodes.

```python
index_cols = Assig_neg_only2.columns.tolist()
index_cols.remove("Neighbourhood")
Final_postcodes = Assig_neg_only2
Final_postcodes = Final_postcodes.groupby(index_cols)["Neighbourhood"].apply(list)
Final_postcodes = Final_postcodes.reset_index()
```

The "Neighbourhood" column has lists inside, need to be converted to strings.

```python
# Here we make a list of strings
N_strings = []
for a in Final_postcodes['Neighbourhood']:
    b =', '.join(a)
    N_strings.append(b)
N_str_pdf = pd.DataFrame(data=N_strings)

#Here we connect it to the original dataframe and drop the column with lists inside.
Df_asstrings = pd.concat([Final_postcodes, N_str_pdf], axis=1, join='inner')
Almost_final= Df_asstrings.drop(labels='Neighbourhood', axis=1)
Almost_final
```

Just some renaiming of columns.

```python
Postcodes = Almost_final.rename({'Postcode': 'PostalCode', 0: 'Neighbourhood'}, axis='columns')
```

Starting with analysis...

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

Find the size of the dataframe, group rows by neighbourhood and find the mean frequency of each category in each neighbourhood.

```python
toronto_onehot.shape
```

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

Find 5 most common venues in each neighbourhood.

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

Put it all into pandas dataframe. Start by sorting venues in descending order.

```python
# Function to sort them
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

New dataframe to display 10 most common venues in each neighbourhood.

```python
import numpy as np # library to handle data in a vectorized manner
num_top_venues = 10

indicators = ['st', 'nd', 'rd']
# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

Starting with clustering.

```python
# Run k-means to get 5 clusters.

from sklearn.cluster import KMeans

# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

New dataframe with top 10 venues and to which cluster neighbourhood belongs.

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = Lonlat_incl

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighbourhood')

toronto_merged.head() # check the last columns!
```

Let´s visualize the clusters.

```python
import matplotlib.cm as cm
import matplotlib.colors as colors


# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], kmeans.labels_):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
